Labelling parts of speech for each word

```python
import nltk
from nltk.corpus import state_union #addresses by various Presidents over last 60/70yrs
from nltk.tokenize import PunktSentenceTokenizer #it is unsupervise machine learning tokenizer, you can train it as well
```

** state_union **  #addresses by various Presidents over last 60/70yrs

** PunktSentenceTokenizer ** #it is unsupervise machine learning tokenizer, you can train it as well

```python
train_text  = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
sample_text[1:300]
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
```

```python
tokenized = custom_sent_tokenizer.tokenize(sample_text)
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print (tagged)
    except Exception as e:
        print(str(e))
```

```python
process_content()
```
