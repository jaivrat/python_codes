```python
#all documents from all categories
documents = [ (list(movie_reviews.words(fileid)), category)
                      for category in movie_reviews.categories()
                      for fileid in movie_reviews.fileids(category)]
```

```python
print(documents[0])
```

##### So a document looks like:

([u'plot', u':', u'two', u'teen', ....], u'neg')

```python
#Just to count positive and negative examples

#Method-1 
counts = {u'pos':0, u'neg':0}
def incCounts(tup):
    global counts
    counts[tup[0]] += tup[1]
    return(0)

[ incCounts(tup) for tup in map(lambda x: (x[1],1), documents)]
print(counts)


#Method-2
class Counter():
    def __init__(self):
        self.counts = {u'pos':0, u'neg':0}
        
    def incr(self, tup):
        self.counts[tup[0]] += tup[1]
    
    def count(self):
        return(self.counts)

counter = Counter()
[ counter.incr(tup) for tup in map(lambda x: (x[1],1), documents)]
print(counter.count())
```

```python
all_words = [w.lower() for w in movie_reviews.words()] #convert to lower case
all_words[1:10]
```

```python
all_words = nltk.FreqDist(all_words)
print (all_words)
```

#### We want to have limit on the amount of words. Right now we have frequency distribution above for all words

```python
all_words.keys()[1:10]
print(all_words.values()[1:10])
```

```python
word_features = list(all_words.keys())[:3000] #We want to use top 3000 words
```

```python
word_features[1:10]
```

```python
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:  #note that word_features contains top 3000 words only(above)
        features[w] = (w in words)
    return(features)

#sample test
find_features(movie_reviews.words("neg/cv000_29416.txt"))
```

```python
featuresets = [(find_features(doc), category) for (doc, category) in documents]
```

```python
training_set = featuresets[:1900] #firs 1900
testing_set  = featuresets[1900:] #rest starting 1900th 
```

```python
training_set[1]
```

#### algo details : https://www.youtube.com/watch?v=pc36aYTP44o

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

#### Accuracy

```python

nltk.classify.accuracy(classifier, testing_set)
```
