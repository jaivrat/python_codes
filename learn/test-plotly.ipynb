##### PIE CHART CREATION

```python
#Expenses

#breakdown each category 
groups = ['Rent','Food','Bills','Miscellaneous']
#create amount
amount = [1000,518,331,277]
#style
colors = ['#d32c58', '#f9b1ee', '#b7f9b1', '#b1f5f9']

trace = go.Pie(labels=groups, values=amount,
               hoverinfo='label+percent', textinfo='value', 
               textfont=dict(size=25),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=3)))

#plot - 
iplot([trace])
```

```python
import pandas as pd

pubg = pd.read_csv("/Users/jvsingh/Downloads/P48-Custom-Data-PUBG.csv")
df_pubg = pubg.apply(pd.to_numeric, errors='ignore')
df_new_pubg = df_pubg.head(10)
df_new_pubg
```

```python


trace = go.Scatter(x = df_new_pubg['solo_RoundsPlayed'], 
                  y = df_new_pubg['solo_Wins'],
                  name='Rounds Won')

layout = go.Layout(title = "PUBG Wind vs Rounds Played",
                  plot_bgcolor = 'rgb(230, 230, 230)',
                  showlegend = True)

fig = go.Figure(data=[trace], layout=layout)

iplot(fig)
```

```python
df_new_pubg = df_pubg.head(50)

trace1 = go.Bar(x = df_new_pubg['player_name'], 
                  y = df_new_pubg['solo_RoundsPlayed'],
                  name='Rounds Played')

trace2 = go.Bar(x = df_new_pubg['player_name'], 
                  y = df_new_pubg['solo_Wins'],
                  name=' Wins')
```

```python
#call date and plots for visualization

layout = go.Layout(barmode = 'group')

fig = go.Figure(data = [trace1, trace2], layout=layout)

iplot(fig, filename='grouped-bar')
```
